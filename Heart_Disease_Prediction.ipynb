# Splitting the dataset into Training set and Test set

```python
X = data.drop(columns='target', axis=1)
y = data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=None, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

# Training Logistic Regression model on Training data

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train, y_train)
```

# Accuracy on Training set

```python
X_train_predict = classifier.predict(X_train)
train_score = accuracy_score(X_train_predict, y_train)
print(train_score)
```

# Accuracy on Test set

```python
y_pred = classifier.predict(X_test)
test_score = accuracy_score(y_pred, y_test)
print(test_score)
```

# Displaying Confusion Matrix

```python
cm = confusion_matrix(y_pred, y_test)
cm
```

# Displaying Test set Results

```python
y_pred = classifier.predict(X_test)

np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.values.reshape(len(y_test),1)),1))
```
